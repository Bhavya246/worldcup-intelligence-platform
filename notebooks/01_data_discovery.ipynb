```python
def determine_winner(row):

    if row["home_score"] > row["away_score"]:
        return row["home_team"]

    elif row["away_score"] > row["home_score"]:
        return row["away_team"]

    else:
        return "Draw"
```

```python
matches_2022["winner"] = matches_2022.apply(
    determine_winner,
    axis = 1
)
```

```python
matches_2022[
    [
        "home_team",
        "away_team",
        "home_score",
        "away_score",
        "winner"
    ]
].head(10)
```

```python
matches_2022["winner"].value_counts()
```

```python
def determine_result_type(row):
    if row["home_score"] > row["away_score"]:
        return "Home Win"

    elif row["away_score"] > row["home_score"]:
        return "Away Win"

    else:
        return "Draw"
```

```python
matches_2022["result_type"] = matches_2022.apply(
    determine_result_type,
    axis = 1
)
```
